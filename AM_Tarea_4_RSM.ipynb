```python
import numpy as np
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
```

```python
# Generar datos de ejemplo
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Añadir el término de sesgo (columna de unos)
X_b = np.c_[np.ones((100, 1)), X]
```

```python
# Formulación básica de la regresión lineal
theta_basic = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
# Formulación de la regresión Ridge (regularización)
lambda_ridge = 1
ridge_reg = Ridge(alpha=lambda_ridge, solver="cholesky")
ridge_reg.fit(X, y)
theta_ridge = np.r_[ridge_reg.intercept_, ridge_reg.coef_.flatten()]
```

```python
# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.plot(X, y, "b.", label="Datos")
plt.plot(X, X_b @ theta_basic, "r-", label="Regresión Básica")
plt.plot(X, X_b @ theta_ridge, "g--", label="Regresión Ridge")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Imprimir coeficientes
print("Coeficientes de la Regresión Básica:", theta_basic.flatten())
print("Coeficientes de la Regresión Ridge:", theta_ridge.flatten())
```
