```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función cuadrática f1
def f1(x1, x2):
    x3 = 3  # Tratamos x3 como una constante
    return 10*x1**2 - x1*x2 - 5*x1*x3 + 5*x1 + 10*x2**2 - 11*x2*x3 - 2*x2 - 5*x3*x1 - 11*x3*x2 - 4*x3**2 + 6*x3 + 9
```

```python
# Definir el gradiente de f1
def grad_f1(x1, x2):
    x3 = 3
    df1_dx1 = 20*x1 - x2 - 5*x3 + 5
    df1_dx2 = -x1 + 20*x2 - 11*x3 - 2
    return np.array([df1_dx1, df1_dx2])
```

```python
# Crear una cuadrícula de valores para x1 y x2
x1_vals = np.linspace(-10, 10, 400)
x2_vals = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
```

```python
# Calcular los valores de f1 en la cuadrícula
Z = f1(X1, X2)
```

```python
# Calcular el gradiente en la cuadrícula
grad = grad_f1(X1, X2)
grad_X1 = grad[0]
grad_X2 = grad[1]
```

```python
# Visualizar la función
plt.figure(figsize=(12, 8))
contour = plt.contour(X1, X2, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.quiver(X1, X2, grad_X1, grad_X2, color='red')
plt.title('Función $f_1(x_1, x_2)$ y su gradiente')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.show()
```
