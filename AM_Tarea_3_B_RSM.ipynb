```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar datos aleatorios para x y m
np.random.seed(0)
N = 100  # Número de muestras
D = 3    # Dimensión del vector x
Q = 3    # Dimensión del vector m

x = np.random.randn(N, D)
m = np.random.randn(N, Q)
```

```python
# Generar pesos aleatorios y términos independientes
w = np.random.randn(D)
w0 = np.random.randn()
p = np.random.randn(Q)
p0 = np.random.randn()
```

```python
# Generar q1 y q2 aleatorios
q1 = np.random.randn()
q2 = np.random.randn()
```

```python
# Calcular z1 y z2
z1 = np.dot(x, w) + w0
z2 = np.dot(m, p) + p0
```

```python
# Generar valores observados y con ruido
y_true = q1 * z1 + q2 * z2
y_observed = y_true + 0.1 * np.random.randn(N)  # Añadir un poco de ruido
```

```python
# Calcular los valores predichos
y_pred = q1 * z1 + q2 * z2
```

```python
# Calcular la función de error (residuos al cuadrado)
errors = (y_observed - y_pred) ** 2
error_total = np.sum(errors)
```

```python
# Imprimir el error total
print(f"Error total: {error_total}")
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(y_observed, y_pred, c=errors, cmap='viridis', alpha=0.7)
plt.colorbar(label='Error Cuadrado')
plt.plot([min(y_observed), max(y_observed)], [min(y_observed), max(y_observed)], 'r--')  # Línea de referencia
plt.xlabel('Valores Observados')
plt.ylabel('Valores Predichos')
plt.title('Gráfico de Dispersión de Valores Observados vs Predichos')
plt.show()
```
