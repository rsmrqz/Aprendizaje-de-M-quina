```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigma_norm(x, Sigma):
    return np.sqrt(np.dot(x.T, np.dot(Sigma, x)))
```

```python
# Definir la matriz Sigma (simétrica y definida no negativa)
Sigma = np.array([[2, 1], [1, 2]])
```

```python
# Generar puntos en un círculo unitario para ilustrar la norma
theta = np.linspace(0, 2 * np.pi, 100)
circle = np.array([np.cos(theta), np.sin(theta)])
```

```python
# Calcular la norma Sigma para cada punto en el círculo
norms = np.apply_along_axis(lambda x: np.dot(Sigma, x), 0, circle)
```

```python
# Plotear el círculo unitario original y el círculo transformado por la norma Sigma
plt.figure(figsize=(8, 8))
plt.plot(circle[0, :], circle[1, :], label='Círculo unitario')
plt.plot(norms[0, :], norms[1, :], label='Transformado por $\Sigma$')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ilustración de la Norma $\Sigma$')
plt.grid(True)
plt.axis('equal')
plt.show()
```
